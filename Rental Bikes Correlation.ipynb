```python
#Import needed libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
pd.options.mode.chained_assignment = None
```

```python
#Read in csv file
df = pd.read_csv('/Users/tracylam/Downloads/RentalBikes.csv')
```

```python
#Seeing the top of the data
df.head()
```

```python
#Getting information about the data
df.info()
```

```python
#Checking for any nulls
df.isnull().sum()
```

```python
#Seeing if there are any outliers
df.boxplot(column=['cnt'])
```

```python
#Getting rid of any duplicate rows
df.drop_duplicates()
```

```python
#Ordering data by highest count of rentals to lowest to see
df.sort_values(by=['cnt'], inplace=False, ascending=False)
```

```python
#Looking at the correlation matrix to see the relationship between the variables
df.corr()
```

```python
#Looking at the correlation matrix with a heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix For Bike Rentals')
```

As seen in the matrix above, it appears that the temp, atemp, and year have the highest correlation with the total amount of rented bicycles. 'Instant' was not counted since it is the index. The 'casual' and 'registered' are also not counted since they are the values that make up the 'cnt' variable.

```python
#Unstacking the matrix to take a look at the correlations better
pd.set_option('display.max_rows',None)
correlation_pairs = correlation_matrix.unstack()
sorted_correlation_pairs = correlation_pairs.sort_values()
high_correlation = sorted_correlation_pairs[(sorted_correlation_pairs) > 0.5]
high_correlation
```

```python
#Creating scatterplot for the temp and cnt variables
plt.scatter(x = df['temp'], y = df['cnt'])
plt.title('Temperature vs Total Rented Bikes Count')
plt.xlabel('Total Rented Bikes Count')
plt.ylabel('Temperature in Celsius')
plt.show()
```

```python
#Creating scatterplot for the atemp and cnt variables
#temperature felt means what the temperature felt like
plt.scatter(x = df['atemp'], y = df['cnt'])
plt.title('Temperature Felt vs Total Rented Bikes Count')
plt.xlabel('Total Rented Bikes Count')
plt.ylabel('Temperature Felt in Celsius')
plt.show()
```
